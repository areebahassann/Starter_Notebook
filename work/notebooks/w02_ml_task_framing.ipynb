## 1. My lane as an ML task (type)


I am continuing with **Lane 2: Refresh / Content Opportunity Scoring**.

The ML task type is primarily a **classification and ranking problem**.

The classification task is identifying whether a content page shows evidence of decline risk using the starter proxy label.

The ranking task is more important from a business perspective because content teams need a prioritized review queue rather than a simple yes/no prediction.

The final output would be a ranked list of pages ordered by their likelihood of needing review.

## 2. Target or proxy


For the starter dataset, I will use the provided proxy label:

is_declining_label = (trend_direction == "down")

This is a proxy rather than an ideal target because it is based on current trend direction instead of a future observed outcome.

A stronger future version of the project could use:

Features from a prior time window → decline or recovery in a future time window

However, the starter dataset provides a useful way to frame the problem and test prioritization methods.

```python
import pandas as pd

df = pd.read_csv("Starter_Notebook/data/raw/content_refresh_anonymized.csv")

df["is_declining_label"] = (
    df["trend_direction"] == "down"
).astype(int)

df[["trend_direction", "is_declining_label"]].head(10)
```

## 3. Success metric


The most useful success metric for this lane is Precision@K.

Content reviewers usually have limited capacity and can only review a fixed number of pages.

Precision@50 measures how many of the top 50 recommended pages are actually positive examples according to the chosen target.

I will also consider:

- Average Precision
- Recall
- ROC AUC

However, Precision@K is the metric most closely aligned with the business decision.

## 4. The unit of analysis, as a real dataframe


The unit of analysis is one content page.

Each row represents a single content item and contains observable signals such as impressions, sessions, CTR, position, age, freshness, and engagement metrics.

The model would generate a score for each page and use that score to prioritize review actions.

```python
cols = [
    "content_id",
    "impressions_90d",
    "sessions_90d",
    "ctr",
    "avg_position",
    "content_age_days",
    "trend_direction"
]

df[cols].head(10)
```

## 5. Why ML beats a fixed rule here


A fixed rule can only evaluate one condition at a time.

For example:

- stale visible page
- low CTR visible page
- declining with demand

However, content performance is influenced by multiple signals simultaneously.

Machine learning can learn patterns across impressions, sessions, CTR, position, freshness, engagement, and content age at the same time.

The goal is not to replace human judgment. Instead, ML can help prioritize pages more effectively than a small set of manually defined rules.

The starter project already provides evidence that a learned model can outperform baseline rules. According to the starter results, the random forest achieved a Precision@50 of 0.740 compared with 0.240 for the baseline rule approach.

This suggests that a learned ranking may identify review opportunities more effectively than a fixed rule alone.
